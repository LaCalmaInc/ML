```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def train_and_generate_submission(train_df, test_df, categorical_columns, output_path='submission.csv'):
    """
    Entrena con train_df y genera predicciones para test_df sin columna target

    Parameters:
    -----------
    train_df : DataFrame
        Dataset de entrenamiento con columna 'income'
    test_df : DataFrame
        Dataset de test sin columna 'income'
    categorical_columns : list
        Lista de columnas categóricas
    output_path : str
        Ruta para guardar el archivo de submission
    """
    # Copia de los dataframes
    train = train_df.copy()
    test = test_df.copy()

    print("Shapes iniciales:")
    print(f"Train shape: {train.shape}")
    print(f"Test shape: {test.shape}")

    # Convertir target a formato numérico
    label_encoder = LabelEncoder()
    train['income'] = label_encoder.fit_transform(train['income'])

    # Separar features y target en train
    X_train = train.drop('income', axis=1)
    y_train = train['income']

    # Crear el preprocesador
    categorical_transformer = Pipeline(steps=[
        ('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore'))
    ])

    preprocessor = ColumnTransformer(
        transformers=[
            ('cat', categorical_transformer, categorical_columns)
        ],
        remainder='passthrough',
        sparse_threshold=0
    )

    # Crear y entrenar el pipeline
    model = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
    ])

    print("\nEntrenando modelo...")
    model.fit(X_train, y_train)

    # Generar predicciones para el test
    print("\nGenerando predicciones...")
    predictions_numeric = model.predict(test)
    predictions = label_encoder.inverse_transform(predictions_numeric)

    # Crear DataFrame de submission
    submission = pd.DataFrame({
        'ID': range(len(predictions)),
        'income': predictions
    })

    # Guardar archivo
    submission.to_csv(output_path, index=False)

    print(f"\nSubmission guardada en {output_path}")
    print("\nPrimeras 5 filas del archivo de submission:")
    print(submission.head())

    print("\nDistribución de predicciones:")
    print(submission['income'].value_counts(normalize=True).round(3))

    return model, submission

# Ejemplo de uso
categorical_columns = [
    'workclass', 'education', 'marital-status', 'occupation',
    'relationship', 'race', 'sex', 'native-country'
]

# Entrenar y generar submission
model, submission = train_and_generate_submission(train_df, test_df, categorical_columns)
```
